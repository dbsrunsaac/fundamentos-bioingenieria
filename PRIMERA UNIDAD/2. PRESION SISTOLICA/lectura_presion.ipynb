```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
presion_csv = 'salazar_presion.csv'
datos = pd.read_csv(presion_csv)

print("Columnas disponibles:", datos.columns)

columna_x = datos['Experiment name:']
columna_y = datos['Sample experiment']
print(len(columna_y))
rango = np.arange(0, 12005, 1)
print(len(rango))
```
